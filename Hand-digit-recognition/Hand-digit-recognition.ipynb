```python
from keras.models import load_model
network = load_model('Trainedmodel.h5')
network.summary()
```

```python
from keras.utils import load_img
from keras.utils import img_to_array
img = load_img('digit0.jpeg', color_mode='grayscale')
img
```

```python
newImg = img_to_array(img)
newImg
```

```python
# Data Preprocessing 
newImg = newImg.reshape((1,28*28))
newImg = newImg.astype('float32') / 255
newImg.shape
```

```python
prediction = network.predict(newImg)
prediction
```

```python
class_names = ['Zero',
               'One',
               'Two',
               'Three',
               'Four',
               'Five',
               'Six',
               'Seven',
               'Eight',
               'Nine',
              ]
```

```python
import numpy as np
np.argmax(prediction)
```

```python
digit = class_names[np.argmax(prediction)]
print(f'Digit:{digit}')
```
